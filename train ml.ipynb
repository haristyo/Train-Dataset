```python

data["Gender"] = data["Gender"].fillna(data["Gender"].mode()[0])
data["Married"] = data["Married"].fillna(data["Married"].mode()[0])
data["Dependents"] = data["Dependents"].fillna(data["Dependents"].mode()[0])
data["Self_Employed"] = data["Self_Employed"].fillna(data["Self_Employed"].mode()[0])
data["LoanAmount"] = data["LoanAmount"].fillna(data["LoanAmount"].mean())
data["Loan_Amount_Term"] = data["Loan_Amount_Term"].fillna(data["Loan_Amount_Term"].mean())
data["Credit_History"] = data["Credit_History"].fillna(data["Credit_History"].mean())
```

melakukan labeling

```python
map_gender = dict(zip(set(data["Gender"]), [1,2]))
data["Gender"] = data["Gender"].apply(lambda x:map_gender[x])
# print(map_gender)

map_edu = dict(zip(set(data['Education']), [1,2]))
data['Education'] = data['Education'].apply(lambda y:map_edu[y])
# print(map_edu)

map_depentdents = dict(zip(set(data['Dependents']), [1,2,3,4]))
data['Dependents'] = data['Dependents'].apply(lambda y:map_depentdents[y])
# print(map_depentdents)

map_propertyarea = dict(zip(set(data['Property_Area']), [1,2,3,4]))
data['Property_Area'] = data['Property_Area'].apply(lambda y:map_propertyarea[y])
# print(map_propertyarea)


map_maried = dict(zip(set(data['Married']), [1,2,3,4]))
data['Married'] = data['Married'].apply(lambda y:map_maried[y])
# print(map_maried)

map_loan_status = dict(zip(set(data['Loan_Status']), [1,2,3,4]))
data['Loan_Status'] = data['Loan_Status'].apply(lambda y:map_loan_status[y])
# print(map_loan_status)

map_self_employed = dict(zip(set(data['Self_Employed']), [1,2,3,4]))
data['Self_Employed'] = data['Self_Employed'].apply(lambda y:map_self_employed[y])
# print(map_loan_status)
data.head()
```
